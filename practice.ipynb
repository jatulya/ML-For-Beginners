```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import model_selection, linear_model, datasets
```

```python
X, y = datasets.load_linnerud(return_X_y=True)
print("Dimensions of X and sample data: ", X.shape, X[0])
print("Dimensions of y and sample data: ", y.shape, y[0])
```

```python
X=X[:,1]
X=X.reshape((-1,1))
y=y[:,1]
y=y.reshape((-1,1))
print("Dimensions of X and sample data: ", X.shape, X[0])
print("Dimensions of y and sample data: ", y.shape, y[0])
```

```python
X_tr, X_te, y_tr, y_te = model_selection.train_test_split(X,y, test_size=0.33)
print("Data is splitted") 
```

```python
model = linear_model.LinearRegression()
print("Model loaded")
```

```python
model.fit(X_tr, y_tr)
print("Model is trained")
```

```python
y_pred = model.predict(X_te)
plt.scatter(X_te, y_te,  color='black')
plt.plot(X_te, y_pred, color='blue', linewidth=3)
plt.xlabel('Waistline')
plt.ylabel('Situps')
plt.title('A Graph Plot Showing Waistline v/s how many situps are needed')
plt.show()
```
